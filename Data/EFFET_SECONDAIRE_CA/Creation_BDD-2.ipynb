#### Tri des colonnes de la bdd

```python
# Supression ds colonnes inutiles
reports_drop = ['REPORT_NO', 'VERSION_NO', 'DATE_INT_RECEIVED', 'NAH_NO','REPORT_TYPE_CODE','REPORT_TYPE_ENG','REPORT_TYPE_FR','GENDER_CODE','GENDER_ENG',
                     'AGE', 'AGE_UNIT_ENG', 'AGE_UNIT_FR', 'OUTCOME_CODE', 'OUTCOME_ENG', 'OUTCOME_FR','WEIGHT_UNIT_ENG', 
                      'HEIGHT_UNIT_ENG','SERIOUSNESS_CODE', 'SERIOUSNESS_ENG', 'REPORTER_TYPE_ENG','REPORTER_TYPE_FR', 'SOURCE_CODE','SOURCE_ENG','SOURCE_FR', 'E2B_IMP_SAFETY_REPORT_ID','AUTHORITY_NUMB','COMPAGNY_NUMB' 
                    ]

reports__indication_drop = ['REPORT_DRUG_ID', 'DRUG_PRODUCT_ID', 'INDICATION_NAME_ENG'] 

df_reports = df_reports.drop(columns=reports_drop, axis=1)
df_reports_indication = df_reports_indication.drop(columns=reports__indication_drop, axis=1)
```

```python
# Supprimer les lignes avec plus de 5 valeurs manquantes
num_columns = df_reports.shape[1]
threshold = num_columns - 5

df_reports = df_reports.dropna(thresh=threshold)
df_reports
```
