```python
pourcentage_manquant = (df_final.isna().mean()) * 100

# Afficher le résultat
print("Pourcentage de données manquantes pour chaque colonne :")
print(pourcentage_manquant)
```

```python
colonnes_a_supprimer = ["date dispo", "date maj","date debut","libelle statut","texte"]

# Créer un nouveau DataFrame en supprimant les colonnes spécifiées
df_final = df_final.drop(columns=colonnes_a_supprimer)
```

```python
len(df_final)
```

```python

nb_valeurs_manquantes_par_ligne = df_final.isna().sum(axis=1)

# Compter le nombre de lignes ayant plus de 90% de valeurs manquantes
nb_lignes_plus_90pc_manquantes = (nb_valeurs_manquantes_par_ligne > 0.9 * len(df_final.columns)).sum()

# Afficher le résultat
print("Nombre de lignes avec plus de 90% de valeurs manquantes :", nb_lignes_plus_90pc_manquantes)
```

```python
lignes_plus_90pc_manquantes = df_final[nb_valeurs_manquantes_par_ligne > 0.9 * len(df_final.columns)]

# Afficher les valeurs de chaque colonne pour ces lignes
print("Valeurs des colonnes pour les lignes avec plus de 90% de valeurs manquantes :")
print(lignes_plus_90pc_manquantes)
```

```python
masque_lignes_plus_90pc_manquantes = nb_valeurs_manquantes_par_ligne > 0.9 * len(df_final.columns)
df_final = df_final.drop(df_final[masque_lignes_plus_90pc_manquantes].index)
```
