#### CIS_bdpm.txt

```python
import pandas as pd 

noms_colonnes = ["CIS","dénomination","Forme","administration","AMM statut","AMM type","Etat","AMM date","StatutBdm","Num autorisation","Titulaire","surveillance"]
cis_bdpm = pd.read_csv("CIS_bdpm.txt", encoding='latin-1', delimiter='\t', names=noms_colonnes)
```

```python
cis_bdpm.head()
```

```python
import matplotlib.pyplot as plt

# Remplacez 'Nom_colonne' par le nom de la colonne pour laquelle vous voulez créer l'histogramme
nom_colonne = 'AMM statut'

# Créer l'histogramme
plt.hist(cis_bdpm[nom_colonne], bins=30, color='blue', alpha=0.7)

# Ajouter des étiquettes et un titre
plt.xlabel('Valeur')
plt.ylabel('Fréquence')
plt.title('Histogramme de la colonne {}'.format(nom_colonne))

plt.xticks(rotation=45)

# Afficher l'histogramme
plt.show()
```
